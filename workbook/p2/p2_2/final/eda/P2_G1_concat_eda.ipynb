```python
from pathlib import Path
import os
import warnings

MPLCONFIG_DIR = Path("/tmp/matplotlib-sbs-datascience")
MPLCONFIG_DIR.mkdir(parents=True, exist_ok=True)
os.environ.setdefault("MPLCONFIGDIR", str(MPLCONFIG_DIR))

import matplotlib as mpl
import matplotlib.font_manager as fm
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns
from IPython.display import Image, Markdown, display

warnings.filterwarnings("ignore", category=UserWarning)
warnings.filterwarnings("ignore", category=FutureWarning)

BASE_DIR = Path("/home/sieg/projects-wsl/SBS_dataScience/workbook/p2/p2_2")
CSV_PATH = BASE_DIR / "final/data/P2_G1_concat.csv"
EDA_DIR = BASE_DIR / "final/eda"
TABLE_DIR = EDA_DIR / "tables"
FIG_DIR = EDA_DIR / "figures"
REPORT_PATH = EDA_DIR / "P2_G1_concat_eda_report.md"
OUTPUT_ENCODING = "utf-8-sig"
CSV_ENCODING_CANDIDATES = ("utf-8-sig", "utf-8", "cp949", "euc-kr")

for path in [EDA_DIR, TABLE_DIR, FIG_DIR]:
    path.mkdir(parents=True, exist_ok=True)

KOREAN_FONT_FILES = [
    Path("/usr/share/fonts/truetype/nanum/NanumGothic.ttf"),
    Path("/usr/share/fonts/opentype/noto/NotoSansCJK-Regular.ttc"),
    Path("/usr/share/fonts/opentype/noto/NotoSansCJKkr-Regular.otf"),
]
FONT_CANDIDATES = ["NanumGothic", "Noto Sans CJK KR", "Malgun Gothic", "AppleGothic", "DejaVu Sans"]


def configure_korean_font() -> str:
    for font_path in KOREAN_FONT_FILES:
        if font_path.exists():
            try:
                fm.fontManager.addfont(str(font_path))
            except RuntimeError:
                pass

    available_fonts = {font.name for font in fm.fontManager.ttflist}
    selected_font = next((font for font in FONT_CANDIDATES if font in available_fonts), "DejaVu Sans")
    mpl.rcParams["font.family"] = [selected_font]
    mpl.rcParams["font.sans-serif"] = [selected_font, *mpl.rcParams.get("font.sans-serif", [])]
    mpl.rcParams["axes.unicode_minus"] = False
    sns.set_theme(style="whitegrid", font=selected_font, rc={"axes.unicode_minus": False})
    return selected_font


def read_csv_with_encoding(path: Path) -> tuple[pd.DataFrame, str]:
    last_error: Exception | None = None
    for encoding in CSV_ENCODING_CANDIDATES:
        try:
            loaded = pd.read_csv(path, encoding=encoding)
        except UnicodeDecodeError as exc:
            last_error = exc
            continue
        loaded.columns = loaded.columns.map(lambda col: str(col).replace("\ufeff", "").strip())
        return loaded, encoding
    raise UnicodeDecodeError(
        "csv",
        b"",
        0,
        1,
        f"failed to read {path} with {CSV_ENCODING_CANDIDATES}: {last_error}",
    )


KOREAN_FONT = configure_korean_font()
df, csv_encoding = read_csv_with_encoding(CSV_PATH)

key_cols = ["기준연도", "학교명", "학과_계열"]
id_cols = ["기준연도", "학교명", "학과_전공", "학과_계열", "학점포기제유무"]
numeric_cols = [
    "입결_프록시",
    "A비율",
    "CD비율",
    "F비율",
    "전체취업률",
    "건보가입취업률",
    "전체진학률",
    "전문대학진학률",
    "대학진학률",
    "대학원진학률",
    "국내진학률",
    "국외진학률",
]
numeric_cols = [col for col in numeric_cols if col in df.columns]

missing_required_cols = [col for col in id_cols if col not in df.columns]
if missing_required_cols:
    raise KeyError(f"CSV 필수 컬럼 누락: {missing_required_cols}")

text_cols = [col for col in id_cols if col in df.columns and col != "기준연도"]
for col in text_cols:
    df[col] = (
        df[col]
        .astype("string")
        .str.replace("\ufeff", "", regex=False)
        .str.strip()
        .replace({"": pd.NA})
    )

for col in numeric_cols:
    df[col] = pd.to_numeric(df[col], errors="coerce")

overview = pd.DataFrame(
    [
        {"metric": "rows", "value": len(df), "note": "1 row = 2024 university-department-label"},
        {"metric": "columns", "value": df.shape[1], "note": "final analytic columns"},
        {"metric": "csv_encoding", "value": csv_encoding, "note": "encoding used by pd.read_csv"},
        {"metric": "matplotlib_font", "value": KOREAN_FONT, "note": "font used for Korean plot text"},
        {"metric": "universities", "value": df["학교명"].nunique(dropna=True), "note": "unique 학교명"},
        {"metric": "department_labels", "value": df["학과_계열"].nunique(dropna=True), "note": "unique normalized labels"},
        {"metric": "raw_department_name_cells", "value": df["학과_전공"].nunique(dropna=True), "note": "unique raw-name bundles"},
        {"metric": "duplicate_key_rows", "value": int(df.duplicated(key_cols, keep=False).sum()), "note": "duplicate 기준연도+학교명+학과_계열 rows"},
        {"metric": "admission_non_null", "value": int(df["입결_프록시"].notna().sum()), "note": "raw_percentile_70cut mapped to final key"},
        {"metric": "employment_non_null", "value": int(df["전체취업률"].notna().sum()), "note": "employment matched to final key"},
        {"metric": "advancement_non_null", "value": int(df["전체진학률"].notna().sum()), "note": "advancement matched to final key"},
        {"metric": "policy_values", "value": ", ".join(map(str, sorted(df["학점포기제유무"].dropna().unique()))), "note": "binary credit-forfeit variable"},
    ]
)
overview.to_csv(TABLE_DIR / "01_overview.csv", index=False, encoding=OUTPUT_ENCODING)

missing = (
    df.isna()
    .sum()
    .rename("missing_n")
    .reset_index()
    .rename(columns={"index": "column"})
)
missing["missing_pct"] = missing["missing_n"] / len(df) * 100
missing["dtype"] = missing["column"].map(lambda col: str(df[col].dtype))
missing = missing.sort_values(["missing_pct", "missing_n"], ascending=False)
missing.to_csv(TABLE_DIR / "02_missing_profile.csv", index=False, encoding=OUTPUT_ENCODING)

numeric_stats = df[numeric_cols].describe(percentiles=[0.01, 0.05, 0.25, 0.5, 0.75, 0.95, 0.99]).T
numeric_stats["missing_n"] = df[numeric_cols].isna().sum()
numeric_stats["missing_pct"] = numeric_stats["missing_n"] / len(df) * 100
numeric_stats["skew"] = df[numeric_cols].skew(numeric_only=True)
numeric_stats["kurtosis"] = df[numeric_cols].kurt(numeric_only=True)
numeric_stats = numeric_stats.reset_index().rename(columns={"index": "column"})
numeric_stats.to_csv(TABLE_DIR / "03_numeric_descriptive_stats.csv", index=False, encoding=OUTPUT_ENCODING)

outlier_rows = []
for col in numeric_cols:
    s = df[col].dropna()
    if s.empty:
        outlier_rows.append(
            {
                "column": col,
                "non_null_n": 0,
                "q1": np.nan,
                "q3": np.nan,
                "iqr": np.nan,
                "lower_bound": np.nan,
                "upper_bound": np.nan,
                "lower_outlier_n": 0,
                "upper_outlier_n": 0,
                "out_of_0_100_n": 0,
                "zero_n": 0,
            }
        )
        continue
    q1, q3 = s.quantile([0.25, 0.75])
    iqr = q3 - q1
    lower = q1 - 1.5 * iqr
    upper = q3 + 1.5 * iqr
    outlier_rows.append(
        {
            "column": col,
            "non_null_n": int(s.size),
            "q1": q1,
            "q3": q3,
            "iqr": iqr,
            "lower_bound": lower,
            "upper_bound": upper,
            "lower_outlier_n": int((s < lower).sum()),
            "upper_outlier_n": int((s > upper).sum()),
            "out_of_0_100_n": int(((s < 0) | (s > 100)).sum()),
            "zero_n": int((s == 0).sum()),
        }
    )
outliers = pd.DataFrame(outlier_rows)
outliers["outlier_n"] = outliers["lower_outlier_n"] + outliers["upper_outlier_n"]
outliers.to_csv(TABLE_DIR / "04_iqr_outlier_profile.csv", index=False, encoding=OUTPUT_ENCODING)

CATEGORY_PALETTE = [
    "#4E79A7",
    "#F28E2B",
    "#59A14F",
    "#E15759",
    "#76B7B2",
    "#EDC948",
    "#B07AA1",
    "#FF9DA7",
    "#9C755F",
    "#BAB0AC",
]
POLICY_ORDER = ["O", "X"]
POLICY_COLOR_MAP = {"O": "#2F80ED", "X": "#F2994A", "결측": "#9CA3AF"}


def make_category_profile(frame: pd.DataFrame, column: str, top_n: int = 30) -> pd.DataFrame:
    counts = (
        frame[column]
        .astype("string")
        .fillna("결측")
        .value_counts(dropna=False)
        .rename_axis("category_value")
        .reset_index(name="row_count")
    )
    if column == "학점포기제유무":
        order_lookup = {value: idx for idx, value in enumerate([*POLICY_ORDER, "결측"])}
        counts["_order"] = counts["category_value"].map(order_lookup).fillna(99)
        counts = counts.sort_values(["_order", "category_value"]).drop(columns="_order")
    else:
        counts = counts.sort_values(["row_count", "category_value"], ascending=[False, True])

    counts = counts.head(top_n).reset_index(drop=True)
    counts.insert(0, "category_rank", np.arange(1, len(counts) + 1))
    counts.insert(0, "category_column", column)
    counts["row_pct"] = counts["row_count"] / len(frame) * 100
    if column == "학점포기제유무":
        counts["color_hex"] = counts["category_value"].map(POLICY_COLOR_MAP).fillna("#9CA3AF")
    else:
        counts["color_hex"] = [CATEGORY_PALETTE[(rank - 1) % len(CATEGORY_PALETTE)] for rank in counts["category_rank"]]
    return counts[["category_column", "category_rank", "category_value", "row_count", "row_pct", "color_hex"]]


categorical_cols = ["기준연도", "학교명", "학과_계열", "학점포기제유무"]
categorical_profile = pd.concat(
    [make_category_profile(df, col, top_n=30) for col in categorical_cols],
    ignore_index=True,
)
categorical_profile.insert(0, "row_no", np.arange(1, len(categorical_profile) + 1))
categorical_profile.to_csv(TABLE_DIR / "05_categorical_top_counts.csv", index=False, encoding=OUTPUT_ENCODING)
category_color_index = categorical_profile[
    ["category_column", "category_rank", "category_value", "color_hex"]
].copy()
category_color_index.to_csv(TABLE_DIR / "15_category_color_index.csv", index=False, encoding=OUTPUT_ENCODING)

duplicate_keys = df[df.duplicated(key_cols, keep=False)].sort_values(key_cols)
duplicate_keys.to_csv(TABLE_DIR / "06_duplicate_key_rows.csv", index=False, encoding=OUTPUT_ENCODING)

pearson = df[numeric_cols].corr(method="pearson")
spearman = df[numeric_cols].corr(method="spearman")
pearson.index.name = "column"
spearman.index.name = "column"
pearson.to_csv(TABLE_DIR / "07_correlation_pearson.csv", index=True, index_label="column", encoding=OUTPUT_ENCODING)
spearman.to_csv(TABLE_DIR / "08_correlation_spearman.csv", index=True, index_label="column", encoding=OUTPUT_ENCODING)

policy_summary = (
    df.groupby("학점포기제유무", dropna=False)[numeric_cols]
    .agg(["count", "mean", "median", "std"])
)
policy_summary.columns = ["_".join(map(str, col)).strip("_") for col in policy_summary.columns.to_flat_index()]
policy_summary = policy_summary.reset_index()
policy_summary.to_csv(TABLE_DIR / "09_policy_group_summary.csv", index=False, encoding=OUTPUT_ENCODING)

univ_coverage = (
    df.groupby("학교명", dropna=False)
    .agg(
        rows=("학과_계열", "count"),
        dept_labels=("학과_계열", "nunique"),
        admission_non_null=("입결_프록시", lambda s: int(s.notna().sum())),
        employment_non_null=("전체취업률", lambda s: int(s.notna().sum())),
        advancement_non_null=("전체진학률", lambda s: int(s.notna().sum())),
        a_rate_mean=("A비율", "mean"),
        employment_rate_mean=("전체취업률", "mean"),
        advancement_rate_mean=("전체진학률", "mean"),
    )
    .reset_index()
)
univ_coverage["admission_coverage_pct"] = univ_coverage["admission_non_null"] / univ_coverage["rows"] * 100
univ_coverage["employment_coverage_pct"] = univ_coverage["employment_non_null"] / univ_coverage["rows"] * 100
univ_coverage["advancement_coverage_pct"] = univ_coverage["advancement_non_null"] / univ_coverage["rows"] * 100
univ_coverage.to_csv(TABLE_DIR / "10_university_coverage_summary.csv", index=False, encoding=OUTPUT_ENCODING)

dept_summary = (
    df.groupby("학과_계열", dropna=False)
    .agg(
        universities=("학교명", "nunique"),
        rows=("학교명", "count"),
        admission_non_null=("입결_프록시", lambda s: int(s.notna().sum())),
        a_rate_mean=("A비율", "mean"),
        cd_rate_mean=("CD비율", "mean"),
        f_rate_mean=("F비율", "mean"),
        employment_rate_mean=("전체취업률", "mean"),
        advancement_rate_mean=("전체진학률", "mean"),
    )
    .reset_index()
    .sort_values(["universities", "rows"], ascending=False)
)
dept_summary.to_csv(TABLE_DIR / "11_department_label_summary.csv", index=False, encoding=OUTPUT_ENCODING)

fig, ax = plt.subplots(figsize=(9, 5))
missing_plot = missing.sort_values("missing_pct", ascending=True)
ax.barh(missing_plot["column"], missing_plot["missing_pct"], color="#4C78A8")
ax.set_title("컬럼별 결측률")
ax.set_xlabel("결측률(%)")
ax.set_ylabel("")
fig.tight_layout()
missing_fig = FIG_DIR / "01_missingness_by_column.png"
fig.savefig(missing_fig, dpi=180, bbox_inches="tight")
plt.close(fig)

plot_cols = numeric_cols
ncols = 3
nrows = int(np.ceil(len(plot_cols) / ncols))
fig, axes = plt.subplots(nrows, ncols, figsize=(15, max(4, nrows * 3.2)))
axes = np.array(axes).reshape(-1)
for ax, col in zip(axes, plot_cols):
    sns.histplot(df[col].dropna(), bins=30, kde=True, ax=ax, color="#4C78A8")
    ax.set_title(col)
    ax.set_xlabel("")
for ax in axes[len(plot_cols):]:
    ax.axis("off")
fig.suptitle("수치형 변수 분포", y=1.01, fontsize=16)
fig.tight_layout()
dist_fig = FIG_DIR / "02_numeric_distributions.png"
fig.savefig(dist_fig, dpi=180, bbox_inches="tight")
plt.close(fig)

fig, axes = plt.subplots(nrows, ncols, figsize=(15, max(4, nrows * 3.0)))
axes = np.array(axes).reshape(-1)
for ax, col in zip(axes, plot_cols):
    sns.boxplot(x=df[col], ax=ax, color="#F58518")
    ax.set_title(col)
    ax.set_xlabel("")
for ax in axes[len(plot_cols):]:
    ax.axis("off")
fig.suptitle("수치형 변수 상자그림", y=1.01, fontsize=16)
fig.tight_layout()
box_fig = FIG_DIR / "03_numeric_boxplots.png"
fig.savefig(box_fig, dpi=180, bbox_inches="tight")
plt.close(fig)

fig, ax = plt.subplots(figsize=(11, 9))
sns.heatmap(spearman, annot=True, fmt=".2f", cmap="vlag", center=0, square=False, ax=ax)
ax.set_title("스피어만 상관계수")
fig.tight_layout()
corr_fig = FIG_DIR / "04_spearman_correlation_heatmap.png"
fig.savefig(corr_fig, dpi=180, bbox_inches="tight")
plt.close(fig)

policy_order = [value for value in POLICY_ORDER if value in set(df["학점포기제유무"].dropna().astype(str))]
policy_palette = {value: POLICY_COLOR_MAP[value] for value in policy_order}
fig, axes = plt.subplots(1, 2, figsize=(12, 5))
sns.boxplot(data=df, x="학점포기제유무", y="A비율", order=policy_order, palette=policy_palette, ax=axes[0])
axes[0].set_title("학점포기제유무별 A비율")
axes[0].set_xlabel("학점포기제유무")
axes[0].set_ylabel("A비율")
sns.boxplot(data=df, x="학점포기제유무", y="전체취업률", order=policy_order, palette=policy_palette, ax=axes[1])
axes[1].set_title("학점포기제유무별 전체취업률")
axes[1].set_xlabel("학점포기제유무")
axes[1].set_ylabel("전체취업률")
fig.tight_layout()
policy_fig = FIG_DIR / "05_policy_group_boxplots.png"
fig.savefig(policy_fig, dpi=180, bbox_inches="tight")
plt.close(fig)

scatter_df = df[["입결_프록시", "A비율", "전체취업률", "전체진학률", "학점포기제유무"]].dropna(subset=["입결_프록시", "A비율"])
fig, ax = plt.subplots(figsize=(7, 5.5))
if len(scatter_df):
    sns.scatterplot(
        data=scatter_df,
        x="입결_프록시",
        y="A비율",
        hue="학점포기제유무",
        hue_order=policy_order,
        palette=policy_palette,
        alpha=0.55,
        s=28,
        ax=ax,
    )
    sns.regplot(data=scatter_df, x="입결_프록시", y="A비율", scatter=False, lowess=True, color="#E45756", ax=ax)
ax.set_title("입결_프록시와 A비율")
ax.set_xlabel("입결_프록시")
ax.set_ylabel("A비율")
fig.tight_layout()
admission_fig = FIG_DIR / "06_admission_proxy_vs_a_rate.png"
fig.savefig(admission_fig, dpi=180, bbox_inches="tight")
plt.close(fig)

dept_top = categorical_profile[categorical_profile["category_column"].eq("학과_계열")].head(15).copy()
policy_counts = categorical_profile[categorical_profile["category_column"].eq("학점포기제유무")].copy()
fig, axes = plt.subplots(1, 2, figsize=(14, 6), gridspec_kw={"width_ratios": [1.8, 1]})
if len(dept_top):
    dept_plot = dept_top.sort_values("row_count", ascending=True)
    axes[0].barh(dept_plot["category_value"], dept_plot["row_count"], color=dept_plot["color_hex"])
    axes[0].set_title("학과_계열 상위 15개 범주")
    axes[0].set_xlabel("행수")
    axes[0].set_ylabel("")
if len(policy_counts):
    axes[1].bar(policy_counts["category_value"], policy_counts["row_count"], color=policy_counts["color_hex"])
    axes[1].set_title("학점포기제유무 범주 색상")
    axes[1].set_xlabel("범주값")
    axes[1].set_ylabel("행수")
fig.tight_layout()
category_fig = FIG_DIR / "07_categorical_color_index.png"
fig.savefig(category_fig, dpi=180, bbox_inches="tight")
plt.close(fig)

def md_table(frame: pd.DataFrame, max_rows: int = 20) -> str:
    view = frame.head(max_rows).copy()
    view = view.astype(object).where(pd.notna(view), "")
    lines = [
        "| " + " | ".join(map(str, view.columns)) + " |",
        "| " + " | ".join(["---"] * len(view.columns)) + " |",
    ]
    for _, row in view.iterrows():
        lines.append("| " + " | ".join(str(v) for v in row.tolist()) + " |")
    return "\n".join(lines)

top_missing = missing[missing["missing_n"] > 0][["column", "missing_n", "missing_pct"]].copy()
top_missing["missing_pct"] = top_missing["missing_pct"].round(2)
stats_view = numeric_stats[["column", "count", "mean", "std", "min", "50%", "max", "missing_pct", "skew"]].copy()
for col in ["mean", "std", "min", "50%", "max", "missing_pct", "skew"]:
    stats_view[col] = pd.to_numeric(stats_view[col], errors="coerce").round(3)
outlier_view = outliers[["column", "non_null_n", "outlier_n", "out_of_0_100_n", "zero_n", "lower_bound", "upper_bound"]].copy()
for col in ["lower_bound", "upper_bound"]:
    outlier_view[col] = pd.to_numeric(outlier_view[col], errors="coerce").round(3)
category_view = categorical_profile[
    categorical_profile["category_column"].isin(["학점포기제유무", "학과_계열"])
][["row_no", "category_column", "category_rank", "category_value", "row_count", "row_pct", "color_hex"]].copy()
category_view["row_pct"] = category_view["row_pct"].round(3)

key_findings = []
key_findings.append(f"- 데이터는 {len(df):,}행, {df.shape[1]}열이며 대학 {df['학교명'].nunique():,}개와 학과 라벨 {df['학과_계열'].nunique():,}개를 포함한다.")
key_findings.append(f"- CSV는 `{csv_encoding}`으로 읽고, 모든 산출 CSV는 Excel 호환을 위해 `{OUTPUT_ENCODING}`로 저장한다.")
key_findings.append(f"- 그래프 한글은 `{KOREAN_FONT}` 폰트로 렌더링하며, 범주형 색상은 `tables/15_category_color_index.csv`에 고정했다.")
key_findings.append(f"- `입결_프록시`는 {df['입결_프록시'].notna().sum():,}행({df['입결_프록시'].notna().mean()*100:.1f}%)만 매칭되어 입결 포함 분석은 표본 축소를 전제로 해야 한다.")
key_findings.append(f"- `전체취업률`은 {df['전체취업률'].notna().sum():,}행({df['전체취업률'].notna().mean()*100:.1f}%), `전체진학률`은 {df['전체진학률'].notna().sum():,}행({df['전체진학률'].notna().mean()*100:.1f}%)에 존재한다.")
key_findings.append(f"- 최종 컬럼 중 0~100 범위를 벗어난 비율형 값은 {int(outliers['out_of_0_100_n'].sum())}개다.")
key_findings.append(f"- 중복 key(`기준연도+학교명+학과_계열`) 행은 {int(overview.loc[overview['metric'].eq('duplicate_key_rows'), 'value'].iloc[0])}개다.")

report = f"""# P2_G1_concat.csv EDA Report

## 1. 데이터 로드

- 입력 CSV: `{CSV_PATH}`
- 읽은 CSV 인코딩: `{csv_encoding}`
- 산출 CSV 인코딩: `{OUTPUT_ENCODING}`
- 그래프 한글 폰트: `{KOREAN_FONT}`
- 출력 폴더: `{EDA_DIR}`
- 분석 단위: `2024년 대학-학과_계열`
- `대학진학률`: 대학 과정 진학자 / 졸업자
- `대학원진학률`: 대학원 진학자 / 졸업자

## 2. 핵심 진단

{chr(10).join(key_findings)}

## 3. 결측치 상위 컬럼

{md_table(top_missing, 20)}

## 4. 기술통계

{md_table(stats_view, 30)}

## 5. IQR 이상치 및 범위 점검

{md_table(outlier_view, 30)}

## 6. 범주 색인 및 색상표

{md_table(category_view, 35)}

## 7. 산출 테이블

- `tables/01_overview.csv`
- `tables/02_missing_profile.csv`
- `tables/03_numeric_descriptive_stats.csv`
- `tables/04_iqr_outlier_profile.csv`
- `tables/05_categorical_top_counts.csv`
- `tables/06_duplicate_key_rows.csv`
- `tables/07_correlation_pearson.csv`
- `tables/08_correlation_spearman.csv`
- `tables/09_policy_group_summary.csv`
- `tables/10_university_coverage_summary.csv`
- `tables/11_department_label_summary.csv`
- `tables/15_category_color_index.csv`

## 8. 산출 그림

- `figures/01_missingness_by_column.png`
- `figures/02_numeric_distributions.png`
- `figures/03_numeric_boxplots.png`
- `figures/04_spearman_correlation_heatmap.png`
- `figures/05_policy_group_boxplots.png`
- `figures/06_admission_proxy_vs_a_rate.png`
- `figures/07_categorical_color_index.png`
"""

REPORT_PATH.write_text(report, encoding="utf-8")

display(Markdown(report))
for image_path in [missing_fig, dist_fig, box_fig, corr_fig, policy_fig, admission_fig, category_fig]:
    display(Markdown(f"### {image_path.name}"))
    display(Image(filename=str(image_path)))
print(f"EDA report saved: {REPORT_PATH}")
```
